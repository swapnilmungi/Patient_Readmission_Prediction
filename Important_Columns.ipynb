```python
#Functions
def get_value_counts(df):
  for col in df:
    if(df[col].dtype == 'object' and col not in {'diag_1,diag_2,diag_3'}):
      print(f" Column Name: {col}\n")
      #print(df[col].value_counts())
      vc = df[col].value_counts()
      for i in range(len(vc)):
        print(f'{vc.index[i]} : {(vc.values[i]/(df.shape[0]/100))}')
        #print(":")
        #print(vc.values[i]/(df.shape[0]/100))

get_value_counts(df)
```
